```python
from transformers import pipeline
pipe = pipeline("text-generation", model="meta-llama/Llama-3.2-1B-Instruct", eos_token_id=pipe.tokenizer.eos_token_id, return_full_text=False)
```

```python
from huggingface_hub import login
login()
```

```python
question1 = "What is an ACE?"
question2 = "What is a volley?"

prompt = "Answer the question\n" + \
"Question: "+ question1 + "\n" + "Answer:"
print(prompt)
```

```python
result = pipe(prompt, max_new_tokens = 50)
print(result)
```
